LIST METHOD
    
append() - to add item at the end of the list
insert() - to add item at any specific position in the list
pop()    - to delete item using its index position
remove() - to delete item using item name 
copy()   - to make a duplicate of list
sort()   - to to sort the list in a particular order
extend() - to combine two list into a single list
clear()  - to delete all items from the list but not the list

```python
print(apps)
```

```python
apps.append("netflix")
print(apps)
```

```python
apps.insert(0, "pinterest")
print(apps)
```

```python
apps.pop()  #by default it delete last item
print(apps)
```

```python
apps.pop(5)
print(apps)
```

```python
apps.remove("phone")
print(apps)
```

```python
copy = apps.copy()
print(copy)
```

```python
copy.clear()
print(copy)
```

```python
print(apps)
```

```python
apps.sort()
print(apps)
```

```python
apps2 = [1,2,3,4,5]
print(apps2)
```

```python
apps.extend(apps2)
print(apps)
```

```python
apps2.extend(apps)
print(apps2)
```
