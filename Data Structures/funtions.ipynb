```python
def even(a):
    if a%2==0:
        print("even number")
    else:
        print("odd number")

even(4)            
```

```python
even = lambda a: print("even") if a%2==0 else print("odd")
even(123456789625374)
```

```python
# to find the number is prime or not

n = int(input("enter any number"))
if n<2:
    print("not primne")
else:
    for i in range(2,n):
        if n%i==0:
            print("not prime")
            break
    else:
        print("prime")
```
