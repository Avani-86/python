```python
#WAP to choose a pizza topping based on user input

topping = input("Choose a pizza topping (paneer, mushroom, or onion):")

if topping == "paneer":
    print("You chose paneer topping")
elif topping == "mushroom":
    print("You chose mushroom topping")
elif topping == "onion":
    print("You chose onion topping")
else:
    print("Invalid topping choice. Please choose from paneer,mushroom,onion")            
```
