```python
file = open('text.txt','r') #open file in read mode
print(file.read())
```

```python
file = open('text.txt','w') #open file i write mode
file.write("Hello Buddy")
print('File written sucessfully')
file = open('text.txt','r') #open file in read mode
print(file.read())
```

```python
file = open ('text.txt','a') #open file in append mode
file.write("\nHello Again")
print('File written sucessfully')
file = open('text.txt','r') #open file in read mode
print(file.read())
```
